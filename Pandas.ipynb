# Podstawowe struktury danych: pd.Series



```python
s = pd.Series(data=[3, 2, 4, 6]) # utworzenie obiektu typu Series
s
```

```python
s = pd.Series(data=[3., 2, 4, 6], index=['a', 'b', 'c', 'd'], name='sample') # jeden z elementów jest typu float, więc wszystkie wartości przyjeły typ float64
s
```

```python
import numpy as np
s = pd.Series(data=[3., np.nan, 4, 6], index=['a', 'b', 'c', 'd'], name='sample') # np.nan - brak danych. Ważne np. w liczeniu średniej. Wartość NaN nie jest brana pod uwagę - nie zaburza wyniku
s
```

```python
price = pd.Series(data={'Apple': 200, 'CD Projekt': 60, 'Amazon': 1900, 'KGHM': np.nan}) # indeksami są domyślnie klucze słownikas
price
```

```python
price['CD Projekt'] # indeksowanie po kluczu
```

```python
price.iloc[1] # indeksowanie numeryczne. Indeksowanie price[1] jest przestarzałe!
```

```python
price.count() # liczba elementów w obiekcie Series
```

```python
price.value_counts(dropna=False) # argument dropna=False uwzględni wartość NaN. Oblicza który element występuje ile razy
```

```python
price.sum()
```

```python
price.mean() # średnia arytmetyczna wszystkich elementów
```

```python
price.min()
```

```python
price.max()
```

```python
price.std()
```

```python
price.describe() # funkcja describe() podaje statystyki zbiorczo. Elementy NaN zostają pominięte do statystyk
```

```python
price.nlargest(1) # metoda wyciąga obiekty o największej wartości (parametr określa ilość obiektów - tutaj jeden, o najwyższej wartości argument). Domyślnie pomija NaN
```

```python
price.nsmallest(1) # metoda wyciąga obiekty o najmniejszej wartości (parametr określa ilość obiektów). Domyślnie pomija NaN
```

```python
price.rank() # metoda rank() porządkuje obiekt Series wg wartości rosnąco
```

```python
price.sort_values() # sortowanie wg wartości. Domyślnie rosnąco. Parametr ascending=False spowoduje sortowanie malejąco
```

```python
price.sort_values(ascending=False)  # Parametr ascending=False spowoduje sortowanie malejąco
```

```python
price_pln = price.apply(lambda x: x * 3.8) # metoda apply() ("zastosuj") za parametr przyjmuje funkcję. Tutaj funkcja lambda przekonwertuje ceny w $ na ceny w PLN.
# Można również zastosować wcześniej zdefiniowaną funkcję!
```

```python
price_pln
```

# Podstawowe struktury danych: pd.DataFrame ("ramka danych")

```python
import pandas as pd
df = pd.DataFrame(data=[12, 12, 32]) # utworzenie obiektu typu DataFrame
df
```

```python
df = pd.DataFrame(data=[12, 12, 32], index=['first', 'second', 'third'], columns=['col_1']) # utworzenie obiektu typu DataFrame
df
```

```python
df = pd.DataFrame(data={'WIG20': ['PKN ORLEN', 'PKO BP'],
                        'mWIG40': ['Amica', 'Playway']}) # utworzenie obiektu typu DataFrame
df
```

```python
df = pd.DataFrame(data=[[10, 12, 13], [23, 12, 10]], index=['first', 'second'], columns=['col_1', 'col_2', 'col_3']) # utworzenie obiektu typu DataFrame
df
```

```python
df.index # zwraca indeksy obiektu DataFrame
```

```python
df.columns # zwraca nazwy kolumn
```

```python
df.values # zwraca wartości obiektu DataFrame
```

```python
df.info() # zwraca informacje o obiekcie DataFrame
```

```python
df.describe() # zwraca statystyki obiektu DataFrame
```

```python
df.describe().T # zwraca statystyki obiektu DataFrame. T - transpozycja
```

```python
df = pd.DataFrame(data=[[10, 12, 13], [23, 12, 10]], index=['first', 'second'], columns=['col_1', 'col_2', 'col_3']) # utworzenie obiektu typu DataFrame
df
```

```python
df.loc['first'] # zwraca wiersz o nazwie 'first'
```

```python
df.loc['first', 'col_2'] # zwraca wartość w wierszu 'first' i kolumnie 'col_2'
```

```python
df.loc[:, 'col_2'] # zwraca wszystkie wiersze z kolumny 'col_2'
```

```python
df.iloc[0] # zwraca wiersz o indeksie [0]
```

```python
df.iloc[0, 1] # zwraca wartość w wierszu o indeksie [0] i kolumnie o indeksie [1]
```
