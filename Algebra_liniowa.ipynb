# Macierz odwrotna

Zrzut ekranu 2025-09-07 135308.png

```python
B = np.linalg.inv(A) # funkcja inverse (inv() ) obliczy macierz odwrotną do zadanej macierzy. Czyli obliczy macierz B, którą pomnożywszy przez zadaną macierz A otrzymamy macierz jednostkową.
B                    # Przy czym mnożąc odwrotnie (AB) również otrzymamy macież jednostkową -> AB = BA = I
```

```python
np.set_printoptions(precision=4, suppress=True)
np.dot(A, B) # sprawdzenie, czy faktycznie pomnożenie macierzy A i B daje macierz jednostkową
```

```python
np.set_printoptions(precision=4, suppress=True)
np.dot(B, A) # odwrócenie mnożenia
```

# Rozwiązywanie układów równań

Zrzut ekranu 2025-09-07 141114.png

Zrzut ekranu 2025-09-07 141142.png

```python
A = np.array([[2, 4], [1, -1]]) # macierz współczynników A
A
```

```python
B = np.array([[10], [-1]]) # macierz wyrazów wolnych B
B
```

```python
A1 = np.linalg.inv(A) # A1 = odwrotność macierzy A
A1
```

```python
X = np.dot(A1, B) # obliczenie macierzy niewiadomych ze wzoru X = A1 * B
X
```

# Funkcje statystyczne

```python
import numpy as np
np.__version__
```

```python
price = np.array([[12.40, 12.80, 11.90, 12.60, 1000],
                  [12.50, 13.00, 11.70, 12.20, 2000],
                  [12.20, 13.40, 12.20, 13.20, 1500]])
print(price)
```

sum()

```python
price.sum() # suma wsyzstkich elementów tablicy
```

```python
np.sum(price) # inna opcja sumowania wszystkich elementów tablicy
```

```python
price.sum(axis=0) # suma poszczególnych kolumn
```

```python
np.sum(price, axis=0) # inna opcja sumowania kolumn
```

```python
price.sum(axis=1) # suma poszczególnych wierszy
```

min()

```python
np.min(price) # wartość minimalna z całej tablicy
```

```python
np.min(price, axis=0) # wartość minimalna z każdej kolumny
```

max()

```python
np.max(price) # wartość maksymalna z całej tablicy
```

```python
np.max(price, axis=0) # wartość maksymalna z każdej kolumny
```

median()

```python
np.median(price) # mediana z całej tablicy. Można liczyć również dla axis= 0 / 1
```

mean()

```python
np.mean(price) # średnia arytmetyczna z całej tablicy. Można liczyć również dla axis= 0 / 1
```

std() - odchylenie standardowe

```python
np.std(price) # odchylenie standardowe z całej tablicy. Można liczyć również dla axis= 0 / 1
```

```python
np.std(price, axis=0)
```

var() - wariancja

```python
np.var(price, axis=0)
```
